# Математическая модель нейрона




*Разделяющая поверхность - это геометрическое место точек, где функция активации меняет свое значение (в любой окрестности любой из таких точек находятся два различных значения функции активации)*

### Формула разделяющей поверхности


*Разделяющая поверхность является подмножеством пространства векторов*

*Пороговая функция активации всегда будет равна 1 с той стороны от разделяющей прямой, в которую указывает вектор w*

*У этого уравнения несчетное (более, чем счетное) множество решений (наборов значений, при которых уравнение является верным равенством) при числе измерений > 1*

*Уравнение задает гиперплоскость (прямую, плоскость, N-мерную плоскость)*

*Вектор w является нормалью к гиперплоскости, которая задана этим уравнением*
